```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
file_path = 'content/API_SP.POP.TOTL_DS2_en_csv_v2_23.csv'
df = pd.read_csv(file_path, encoding='utf-8', skiprows=4)
df.head()
```

```python
df_final = df.drop(['Country Code',	'Indicator Name',	'Indicator Code'], axis=1)
```

```python
# Choose a specific country
chosen_country = 'United States'

# Filter data for the chosen country
country_data = df[df['Country Name'] == chosen_country]

# Ensure columns are numeric and handle missing values
years = country_data.columns[4:-1].astype(str)
population = country_data.iloc[0, 4:-1].values
population = pd.to_numeric(population, errors='coerce')

# Plot the bar chart
plt.figure(figsize=(10, 6))
plt.bar(years, population, color='skyblue')
plt.xlabel('Year')
plt.ylabel('Population')
plt.title(f'Population of {chosen_country} Over the Years')
plt.xticks(rotation=90)
plt.show()
```

```python
plt.hist(population, bins=20, color='skyblue', edgecolor='black')
plt.show()
```
